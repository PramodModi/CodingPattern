## Concept of Monotonic Stack

    A Monotonic Stack is a special kind of stack, which maintains its elements in specific order (increasing or decreasing).
    
    
### Two types of Monotonic Stack:

   #### Increasing Monotonic Stack
    A Monotonic stack is called increasing monotonic stack, if the elements in stack is in increasing order from bottom to top. Here every new element should be greater than or equal to top of stack, if new element is smaller, then pop the element from stack, till top element is smaller or equal to new element or stack is empty.
    
    Ex: array = [1,7,9,5,4,8]
    
    Monotonic stack, while pushing new element:
        MS = [1,7,9] pused, 1,7 and 9 as they are in increasing order
        Push 5. 5 is smaller than top element so remove 9 and 7
        MS = [1,5]
        Push 4. 4 is smaller than 5, so remove 5
        MS = [1,4]
        Push 8.
        MS = [1,4,8]
        
   ##### Template
   
               create an empty stack
                for each element in the array:
                    while stack is not empty AND top of stack is more than the current element:
                        pop the stack
                    push the current element to stack
    
  #### Decreasing Monotonic Stack
    A Monotonic stack is called decreasing monotonic stack, if the elements in stack is in decreasing order from bottom to top. Here every new element should be smaller than or equal to top of stack, if new element is greater, then pop the element from stack, till top element is greater or equal to new element or stack is empty.
    
    Ex: array = [6,3,1,5,6,2]
    
    Monotonic stack, while pushing new element:
        MS = [6,3,1] pused, 6,3,1 as they are in decreasing order
        Push 5. 5 is greater than top element so remove 1 and 3
        MS = [6,5]
        Push 6. 6 is greater than 5, so remove 5
        MS = [6,6]
        Push 2.
        MS = [6,6,2]
        
   ##### Template       
   
                create an empty stack
                    for each element in the array:
                        while stack is not empty AND top of stack is less than the current element:
                            pop the stack
                        push the current element to stack
