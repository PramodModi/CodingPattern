## Problem: Unique Generalised Abbreviations

Given a word, write a function to generate all of its unique generalized abbreviations.
A generalized abbreviation of a word can be generated by replacing each substring of the word with the count of characters in the substring. 

Example1:

    string = "ab"
    output : ['2', '1b', 'a1', 'ab']
    Explanation: Take the example of “ab” which has four substrings: “”, “a”, “b”, and “ab”. After replacing these substrings in the actual word by the count of characters, we get all the generalized abbreviations: “ab”, “1b”, “a1”, and “2”.
    
Example2:

    string = "abc"
    output : ['3', '2c', '1b1', '1bc', 'a2', 'a1c', 'ab1', 'abc']
    


### Approach:

This is simlar to Parenthesis generation problem. In place of number of open and close bracket, we will use index and count.
There are two recurssive methods, one for count and another for characters. 
Result will get appended, once index is equal to length of string. If count is greater than zero then add the count in the temp, before appending to the result, else append the temp only.
As second recursion method is for charecters, so pass count = 0, so that adding count should be avoided before appending the temp in rsult.

Temp will keep updating at each iteration of scond recursion method, with count and charecter at the index in the string.

```python
def findAbbr(string):
    # Write your code here.
    result = []
    helper(string, result, 0, 0,"")
    return result

def helper(string, result, index, count, temp):
    if index == len(string):
        if count > 0:
            temp += str(count)
        result.append(temp)
    else:
        helper(string, result, index+1, count+1, temp)
        w = str(count) if count > 0  else ""
        helper(string, result, index+1, 0, temp+w+string[index] )
```
