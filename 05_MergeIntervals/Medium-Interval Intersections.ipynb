## Problem: Interval Intersections
LeetCode: 986. Interval List Intersections

https://leetcode.com/problems/interval-list-intersections/

You are given two lists of closed intervals, firstList and secondList, where firstList[i] = [starti, endi] and secondList[j] = [startj, endj]. Each list of intervals is pairwise disjoint and in sorted order.

Return the intersection of these two interval lists.

A closed interval [a, b] (with a <= b) denotes the set of real numbers x with a <= x <= b.

The intersection of two closed intervals is a set of real numbers that are either empty or represented as a closed interval. For example, the intersection of [1, 3] and [2, 4] is [2, 3].

 

Example 1:



    Input: firstList = [[0,2],[5,10],[13,23],[24,25]], secondList = [[1,5],[8,12],[15,24],[25,26]]
    Output: [[1,2],[5,5],[8,10],[15,23],[24,24],[25,25]]
Example 2:

    Input: firstList = [[1,3],[5,9]], secondList = []
    Output: []
 

Constraints:

    0 <= firstList.length, secondList.length <= 1000
    firstList.length + secondList.length >= 1
    0 <= starti < endi <= 109
    endi < starti+1
    0 <= startj < endj <= 109 
    endj < startj+1

### Approach :
Given, both lists are in sorted order.
Need to find the overlap. 
Overlap: Suppose a and b are two intervals one from each list. Both are overlapped if a.start <= b.end and b.start <= a.end.
Take the intersection of interval and store it. 
Intersected part: 
    newInterval.start = max(a.start, b.start)
    newInterval.end = min(a.end, b.end)
    
setps:
1. take two iterator, i and j. Both will start from 0 position and run till the end of lists.
2. Check the overlap condition. If overlap take the intersected part and store it
3. Check which one is smaller, so that next interval can be taken from that list.
    If a.end < b.end, means need to take next interval from the list whom 'a' is belongs to. As b can be overlapped with next interval also. If not, then take next interval from the list whom 'b' is belongs to.


```python
def intervalIntersection(firstList, secondList):
    if len(firstList) == 0 or len(secondList) == 0:
        return []
    result = []
    i, j = 0, 0
    while i < len(firstList) and j < len(secondList):
        if firstList[i][0] <= secondList[j][1] and secondList[j][0] <= firstList[i][1]:
            start = max(firstList[i][0], secondList[j][0])
            end = min(firstList[i][1] ,secondList[j][1])
            result.append([start, end])
        if firstList[i][1] < secondList[j][1]:
            i += 1
        else:
            j += 1
    return result
```

```python
firstList = [[0,2],[5,10],[13,23],[24,25]]
secondList = [[1,5],[8,12],[15,24],[25,26]]
intervalIntersection(firstList, secondList)
```
