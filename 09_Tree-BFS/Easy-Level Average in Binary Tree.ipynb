## Problem: Level Average in Binary Tree
637. Average of Levels in Binary Tree

https://leetcode.com/problems/average-of-levels-in-binary-tree/

Given the root of a binary tree, return the average value of the nodes on each level in the form of an array. Answers within 10-5 of the actual answer will be accepted.
 

Example 1:



    Input: root = [3,9,20,null,null,15,7]
    Output: [3.00000,14.50000,11.00000]
    Explanation: The average value of nodes on level 0 is 3, on level 1 is 14.5, and on level 2 is 11.
    Hence return [3, 14.5, 11].

Example 2:


    Input: root = [3,9,20,15,7]
    Output: [3.00000,14.50000,11.00000]
 

Constraints:

    The number of nodes in the tree is in the range [1, 104].
    -231 <= Node.val <= 231 - 1

### Approach:
Since this problem is asking to calculate average of each level. We use "BFS Pattern". For more details: https://github.com/PramodModi/CodingPattern/blob/main/09_Tree-BFS/01-Concept%20of%20BFS%20(Level%20Order%20Traversal).ipynb

Will use queue to track of each level.

Steps:

1. Start with root. Enqueue root in the queue
2. Run the loop till queue is empty
3. Initialize a variable say, sum. This variable will have summation of values of nodes at each level. Also keep n = number of nodes in the queue.
3. Run anothe loop till the queue size. Here queue size is nothing but number of node in a particulr level.
4. Dequeue the node from queue, add the value in a variable sum.
5. Enqueue the left child node and right child node, if they are available.
6. Once internal loop is finished means, all nodes of a particular level is exhausted, use sum for calculating the average.
7. Append the rsult with average
8. At the end Return the result.

```python
from collections import deque

class Node:
    def __init__(self, val, left, right):
        self.val = val
        self.left = left
        self.right = right
        
def averageOfLevels(root):
    average = []
    q = deque([root])
    while q:
        n = len(q)
        sum = 0
        for _ in range(n):
            node = q.popleft()
            sum += node.val
            if node.left: q.append(node.left)
            if node.right: q.append(node.right)
        average.append(sum/n)
    return average
    
```

```python
root = Node(3,Node(9, None, None), Node(20, Node(15, None, None), Node(7, None, None)))
averageOfLevels(root)
```
