## Problem: Right View of Binary Tree

LeetCode: 199. Binary Tree Right Side View

https://leetcode.com/problems/binary-tree-right-side-view/

Given the root of a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

 

Example 1:



    Input: root = [1,2,3,null,5,null,4]
    Output: [1,3,4]
Example 2:

    Input: root = [1,null,3]
    Output: [1,3]
Example 3:

    Input: root = []
    Output: []
 

Constraints:

    The number of nodes in the tree is in the range [0, 100].
    -100 <= Node.val <= 100

### Approach:

This problem follows the Binary Tree Level Order Traversal pattern. We can follow the same BFS approach. For more details: https://github.com/PramodModi/CodingPattern/blob/main/09_Tree-BFS/01-Concept%20of%20BFS%20(Level%20Order%20Traversal).ipynb

The only difference will be that while traversing we will remember the node value. At end of each level loop, this value will have value of right node at that particular level. That is the right side view.

Steps:

1. Start with root. Enqueue root in the queue
2. Run the loop till queue is empty
3. Run anothe loop (level order loop) till the queue size. Here queue size is nothing but number of node in a particulr level.
4. Dequeue the node from queue, update the value of node in variable value.
5. Enqueue the left child node and right child node, if they are available.
6. Once internal loop is finished means, all nodes of a particular level is exhausted, append the result with value.
7. At the end Return the result.

```python
from collections import deque

class Node:
    def __init__(self, val, left= None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
def rightSideView(root):
    result = []
    if root is None:
        return result
    q = deque([root])
    while q:
        for _ in range(len(q)):
            node = q.popleft()
            value = node.val
            if node.left: q.append(node.left)
            if node.right: q.append(node.right)
        result.append(value)
    return result
                
    
```

```python
root = Node(1, Node(2,None, Node(5)), Node(3, None, Node(4)))
rightSideView(root)
```

```python
root = Node(1,None, Node(3))
rightSideView(root)
```
