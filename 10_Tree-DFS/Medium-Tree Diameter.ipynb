## Problem: Tree Diameter
LeetCode: 543. Diameter of Binary Tree

https://leetcode.com/problems/diameter-of-binary-tree/

Given the root of a binary tree, return the length of the diameter of the tree.

The diameter of a binary tree is the length of the longest path between any two nodes in a tree. This path may or may not pass through the root.

The length of a path between two nodes is represented by the number of edges between them.

 

Example 1:



    Input: root = [1,2,3,4,5]
    Output: 3
    Explanation: 3 is the length of the path [4,2,1,3] or [5,2,1,3].
Example 2:

    Input: root = [1,2]
    Output: 1
 

Constraints:

    The number of nodes in the tree is in the range [1, 104].
    -100 <= Node.val <= 100

### Approach:

This problem belongs to Tree-DFS Pattern. Here is the slight differences:
At every step, we have to find the height of left and right child of current node. FOr this we will have to make two recursive call.
Diameter of curent node = height of left node + height of right node.
At any step, maximum diameter = maximum of prev caculated diameter and current node diameter. We will store the max diameter in class level variable.
Height of a node = maximum of left node height and right node height + 1  for the current node.


```python
class Node:
    def __init__(self, val,left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
class Info:
    def __init__(self):
        self.diameter = 0
        
def diameterOfBinaryTree(root):
    info = Info()
    calculateHeight(root, info)
    return info.diameter
def calculateHeight(root, info):
    if root is None:
        return 0
    leftHeight = calculateHeight(root.left, info)
    rightHeight = calculateHeight(root.right, info)
    diameter = leftHeight + rightHeight
    maxD = max(diameter, info.diameter)
    info.diameter = maxD
    return max(leftHeight, rightHeight) +1
    
```

```python
root = Node(1, Node(2, Node(4), Node(5)), Node(3))
diameterOfBinaryTree(root)
```

```python
root = Node(1, Node(2))
diameterOfBinaryTree(root)
```
