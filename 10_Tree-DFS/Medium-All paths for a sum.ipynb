## Problem: All paths for a sum
LeetCode : 113. Path Sum II

https://leetcode.com/problems/path-sum-ii/

Given the root of a binary tree and an integer targetSum, return all root-to-leaf paths where the sum of the node values in the path equals targetSum. Each path should be returned as a list of the node values, not node references.

A root-to-leaf path is a path starting from the root and ending at any leaf node. A leaf is a node with no children.

 

Example 1:



    Input: root = [5,4,8,11,null,13,4,7,2,null,null,5,1], targetSum = 22
    Output: [[5,4,11,2],[5,8,4,5]]
    Explanation: There are two paths whose sum equals targetSum:
    5 + 4 + 11 + 2 = 22
    5 + 8 + 4 + 5 = 22
Example 2:


    Input: root = [1,2,3], targetSum = 5
    Output: []
Example 3:

    Input: root = [1,2], targetSum = 0
    Output: []
 

Constraints:

The number of nodes in the tree is in the range [0, 5000].
    -1000 <= Node.val <= 1000
    -1000 <= targetSum <= 1000

### Approach:

This problem can be solved by Tree-DFS pattern. For more details: https://github.com/PramodModi/CodingPattern/blob/main/10_Tree-DFS/01-Concept%20of%20DFS%20(Depth%20First%20Search).ipynb

Here, we need to print the path as well, so will pass result array to store the arrays of values, combined give target sum. Also, need to pass a temp list which will have value of visited node. Once target sum is zero, and it is leaf node, then append the temp values to the result array.

Do not forget to backtrack the temp array at the end of the recurssive call.

```python
class Node:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
def pathSum(root, targetSum):
    result = []
    getPathSum(root, targetSum, result, [])
    return result
def getPathSum(root, targetSum, result, temp):
    if root is None:
        return 
    targetSum -= root.val
    temp.append(root.val)
    if targetSum == 0 and root.left == None and root.right == None:
        result.append(temp[:])
    getPathSum(root.left, targetSum, result, temp)
    getPathSum(root.right, targetSum, result, temp)
    temp.pop()
     
```

```python
root = Node(5, Node(4, Node(11, Node(7), Node(2))), Node(8, Node(13), Node(4, Node(5), Node(1))))
pathSum(root, 22)
```

```python
root = Node(1, Node(2), Node(3))
pathSum(root, 5)
```
