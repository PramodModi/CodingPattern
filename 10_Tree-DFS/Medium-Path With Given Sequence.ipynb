## Problem: Path With Given Sequence

Given a binary tree and a number sequence, find if the sequence is present as a root-to-leaf path in the given tree.

Example:
Given Tree :
    
   
    Input : arr = [5, 2, 4, 8] for above tree
    Output: True
    
    Input : arr = [5, 3, 4, 9] for above tree
    Output: False

### Approach:

This problem can be solved by Tree-DFS pattern. For more details: https://github.com/PramodModi/CodingPattern/blob/main/10_Tree-DFS/01-Concept%20of%20DFS%20(Depth%20First%20Search).ipynb

Steps:

Compare the node value with th array value, since we will be using recursive method, so better to start from last index of array.
In recursive method, if index > len(array), so return False also if number at index in array is not equal to current node value, then return False.
If nidex  == len(arra) -1, that is index is the last position in the array and if number at index in array is equal to the current node value then return True.
Keept iterating for left and right child. Index will keep increamenting by one on each iteration.

```python
class Node:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
def hasSequence(root, arr):
    if root is None:
        return len(arr) == 0
        
    return findSeq(root, arr, 0)

def findSeq(root, arr, index):
    if root is None:
        return False
    if index >= len(arr) or arr[index] != root.val:
        return False
    if index == len(arr)-1 and root.left is None and root.right is None:
        return True
    return findSeq(root.left, arr, index+1) or findSeq(root.right, arr, index+1)
    
```

```python
root = Node(5, Node(2, Node(1), Node(4, Node(6), Node(8))), Node(3))
arr = [5, 2, 4, 8]
hasSequence(root,arr )
```

```python
arr = [5, 3, 4, 9] 
hasSequence(root,arr )
```
