## Problem: Merge K Sorted Lists
LeetCode: 23. Merge k Sorted Lists

https://leetcode.com/problems/merge-k-sorted-lists/

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

 

Example 1:

    Input: lists = [[1,4,5],[1,3,4],[2,6]]
    Output: [1,1,2,3,4,4,5,6]
    Explanation: The linked-lists are:
    [
      1->4->5,
      1->3->4,
      2->6
    ]
    merging them into one sorted list:
    1->1->2->3->4->4->5->6
Example 2:

    Input: lists = []
    Output: []
    Example 3:

    Input: lists = [[]]
    Output: []
 

Constraints:

    k == lists.length
    0 <= k <= 104
    0 <= lists[i].length <= 500
    -104 <= lists[i][j] <= 104
    lists[i] is sorted in ascending order.
    The sum of lists[i].length will not exceed 104.

### Approach:
This problem belongs to K-way merge.
First push first node.value from all lists in the min heap. While pushing the node.val, push node and index of list in lists into minHeap. So that we can take the next node from current list with the help of node or move to next list with the help of index.

Initialize head and temp as None. Head will point to new sorted list and temp will be used to move to the next element and add in the new sorted list.

Run while loop till heap is having data.
Pop the minimum value node from heap. Assign it to head and temp, if it is the first element in sorted list, otherwise assign it to temp.next and move temp to node. Push the next node from current list in to heap, however first check if next node is not None, if it is None then pop the next element from heap and repeat the same.

Return the head.


```python
## This is just for testing purpose. 
class Node:
    def __init__(self, val, nextNode = None):
        self.val = val
        self.next = nextNode


def printList(head):
    print("Priting List")
    temp = head
    while temp:
        print(temp.val)
        temp = temp.next
        
        
## This is the algorithm for this problem
import heapq
def mergeKLists(lists):
    minHeap = []
    for i in range (len(lists)):
        node = lists[i]
        heapq.heappush(minHeap, (node.val, node, i))
    head = None
    temp = None
    
    while minHeap:
        val, node, index = heapq.heappop(minHeap)
        if head is None:
            head = node
            temp = node
        else:
            temp.next = node
            temp = node
        if node.next is None:
            continue
        else:
            node = node.next
            heapq.heappush(minHeap,(node.val, node, index))
    return head
        

        
```

```python
head1 = Node(1, Node(3, Node(5, Node(7))))
printList(head1)
head2 = Node(2, Node(4, Node(6, Node(8))))
printList(head2)

lists = [head1, head2]
head = mergeKLists(lists)

print("Merged List")
printList(head)
```
