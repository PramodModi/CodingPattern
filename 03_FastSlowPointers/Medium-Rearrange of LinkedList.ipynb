## Problem: Rearrange of linkedlist
LeetCode: 143. Reorder List

https://leetcode.com/problems/reorder-list/description/

You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln
Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:


    
    Input: head = [1,2,3,4]
    Output: [1,4,2,3]

Example 2:


    Input: head = [1,2,3,4,5]
    Output: [1,5,2,4,3]
 

Constraints:

    The number of nodes in the list is in the range [1, 5 * 104].
    1 <= Node.val <= 1000

```python
# This is to create a link list
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

def insertNode(head, val):
    newNode = Node(val)
    if head == None:
        head = newNode
        return head
    temp = head
    while temp.next != None:
        temp = temp.next
    temp.next = newNode
    return head

def printList(head):
    temp = head
    while temp:
        print(temp.val, end =" ")
        temp = temp.next
```

```python
def reorderList(head):
    if head.next is None:
        return head
    first = head
    second = splitListinMid(head)
    second = reverse(second)
    newHead = interweaveList(first, second)
    return newHead

def splitListinMid(head):
    if head.next is None:
        return head
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    midHead = slow.next
    slow.next = None
    return midHead

def reverse(head):
    cur = head
    prev = None
    while cur:
        temp = cur
        cur = cur.next
        temp.next = prev
        prev = temp
    return prev

def interweaveList(first, second):
    head = first
    while first and second:
        temp = first.next
        first.next = second
        first = temp
        temp = second.next
        second.next = first
        second = temp
    return head
```

```python
## Example 1
#Create linkedlist

head = None
head = insertNode(head,1)
head = insertNode(head,2)
head = insertNode(head,3)
head = insertNode(head,4)
head = insertNode(head,5)
printList(head)

# Rearrange the list
reorderList(head)
print("\nAfter rearrange")
printList(head)

```
